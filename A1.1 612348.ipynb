**4.** En el siguiente bloque cree la variable **solo_H** para separar la población masculina mediante el uso de la condición == "*Male*" en la columna de Sexo de la hoja de datos.

Después se utiliza para el **eje Y** el nivel de obesidad como en el punto anterior y para el **eje X** se utiliza la altura, una variable cuantitativa que puede conseguirse de la base de datos.

Ya para graficar se utilizó la librería plt. Se utilizó tanto **plt.figure** para definir el tamaño del gráfico, la función **plt.scatter** grafica los datos indicados y con **plt.show()** nos muestra el gráfico.

```python
%matplotlib inline

import matplotlib.pyplot as plt

solo_H = data[data['Sexo'] == "Male"]
plt.figure(figsize = (10,4))
plt.scatter(x = solo_H.Estatura, y = solo_H.NivelDeObesidad, alpha = 0.1)
plt.show()
```
